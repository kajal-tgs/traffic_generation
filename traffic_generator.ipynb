```python
#TO SCRAPE THE WHOLE PROXY LIST FROM FREE PROXY WEBSITE;-
import requests
import random
from bs4 import BeautifulSoup as bs

def get_free_proxies():
    url = "https://free-proxy-list.net/"
    response = requests.get(url)
    if response.status_code != 200:
        print("Failed to retrieve the page.")
        return []

    soup = bs(response.content, 'html.parser')
    table = soup.find("section", attrs={"id": "list"})
    if table is None:
        print("Table not found on the page.")
        return []

    proxies = []
    rows = table.find_all("tr")[1:]
    for row in rows:
        tds = row.find_all("td")
        if len(tds) >= 2:
            ip = tds[0].text.strip()
            port = tds[1].text.strip()
            proxies.append(f"{ip}:{port}")

    return proxies

if __name__ == "__main__":
    proxies = get_free_proxies()
    if proxies:
        print("Free Proxies:")
        for proxy in proxies:
            print(proxy)
    else:
        print("No proxies were retrieved.")
```

```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
def crawl_website_page_with_selenium(ip_address, url):
    firefox_options = webdriver.FirefoxOptions()
    firefox_options.set_preference("network.proxy.type", 1)
    firefox_options.set_preference("network.proxy.http", ip_address)
    firefox_options.set_preference("network.proxy.http_port", 80)
    driver = webdriver.Firefox(options=firefox_options)
    driver.get(url)
    return driver 
def search_case_studies(driver, query):
    search_box = driver.find_element(By.ID, 'appx-search-01')
    search_box.send_keys(query)
    search_box.send_keys(Keys.RETURN)
    content = driver.page_source
    return "Veruna - Insurance Agency Management" in content
ip_address = '23.40.209.159'
url = 'https://appexchange.salesforce.com/appxConsultingListingDetail?listingId=a0N3A00000DmgW2UAJ&legacy=true'
driver = crawl_website_page_with_selenium(ip_address, url)
search_query = 'Veruna - Insurance Agency Management'
result = search_case_studies(driver, search_query)
time.sleep(10)
driver.quit()
```
